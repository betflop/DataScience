````
n_train = 150        
n_test = 1000       
noise = 0.1

def f(x):
    x = x.ravel()
    return np.exp(-x ** 2) + 1.5 * np.exp(-(x - 2) ** 2)

def generate(n_samples, noise):
    X = np.random.rand(n_samples) * 10 - 5 # создаем массив рандомных значений
    X = np.sort(X).ravel() 
    y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2) + \
    np.random.normal(0.0, noise, n_samples) # добавляем шум
    X = X.reshape((n_samples, 1))
    return X, y

X_train, y_train = generate(n_samples=n_train, noise=noise)
X_test, y_test = generate(n_samples=n_test, noise=noise)


from sklearn.tree import DecisionTreeRegressor

reg_tree = DecisionTreeRegressor(max_depth=5, random_state=17)

# print(X_train,type(X_train))
reg_tree.fit(X_train, y_train)
reg_tree_pred = reg_tree.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "b") # Тестовая выборка с значениями без шума
plt.scatter(X_train, y_train, c="b", s=20) # Тренировочная выборка с шумом
plt.plot(X_test, reg_tree_pred, "g", lw=2) # Предсказание
plt.xlim([-5, 5])
plt.title("Decision tree regressor, MSE = %.2f" % np.sum((y_test - reg_tree_pred) ** 2))
plt.show()
````

````
def form_linearly_separable_data(n=500, x1_min=0, x1_max=30, x2_min=0, x2_max=30):
    data, target = [], []
    for i in range(n):
        x1, x2 = np.random.randint(x1_min, x1_max), np.random.randint(x2_min, x2_max)

        # Число по модулю
        if np.abs(x1 - x2) > 0.5:
            data.append([x1, x2])
            # Знак числа 1 или -1
            target.append(np.sign(x1 - x2))
    return np.array(data), np.array(target)

X, y = form_linearly_separable_data()

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='autumn', edgecolors='black');
````

The role of max_features

max_features can be set to any value, and it won't hurt accuracy in our setting. Sklearn's documentation says

Note: the search for a split does not stop until at least one valid partition of the node samples is found, even if it requires to effectively inspect more than max_features features.

So at each split, only one feature will be considered (each time a random one). If a split can not be done with some feature (eg. all values are equal), then another feature is sampled.

So in case of max_features=1 we get a terribly overfitted tree, but in principle the model can fit to the designed y.

````
tree_max_1_feature = DecisionTreeClassifier(max_features=1, random_state=17)
tree_max_1_feature.fit(X, y)
tree_max_1_feature.score(X, y)
````
