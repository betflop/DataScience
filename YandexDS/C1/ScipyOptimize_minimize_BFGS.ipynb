```python
import numpy as np
import math
```

```python
# fx = lambda x: math.sin(x/5) * math.exp(x/10) + 5 * math.exp(-x/2)

def fx(x):
    return math.sin(x/5.) * math.exp(x/10.) + 5 * math.exp(-x/2.)
```

```python
from scipy.optimize import minimize
```

```python
x0 = np.array([1, 1])
```

```python
# res = minimize(fx, x0, method='Nelder-Mead', tol=1e-6)
res = minimize(fx, [2.], method='BFGS')
print res
```

```python
res = minimize(fx, [30.], method='BFGS')
print res
```

```python
from scipy.optimize import differential_evolution
```

```python
differential_evolution(fx,[(1, 30)])
```
