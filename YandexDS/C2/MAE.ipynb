# Разберемся с среднеабсолютной ошибкой

### Для среднеквадратичной мы писали
Напишем функцию, которая по двум параметрам $w_0$ и $w_1$ вычисляет квадратичную ошибку приближения зависимости роста $y$ от веса $x$ прямой линией $y = w_0 + w_1 * x$

$$error(w_0, w_1) = \sum_{i=1}^n {(y_i - (w_0 + w_1 * x_i))}^2 $$

Тоесть параметра 2: w0, w1

Производная для такого случая будет иметь вид (чтобы получить **средне**квадратичную делим на количество элементов)

$$\frac{dQ}{dw_1}=\frac{2}{l}\sum_{i=1}^n {(y_i - (w_0 + w_1 * x_i))*x_i}$$

Для w0

$$\frac{dQ}{dw_0}=\frac{2}{l}\sum_{i=1}^n {(y_i - (w_0 + w_1 * x_i))}$$

Теперь попробуем вычислить MAE ошибку

$$error(w_0, w_1) = \sum_{i=1}^n {|y_i - (w_0 + w_1 * x_i)|} $$

Производная модуля. Воспользуемся определением модуля

$$|f(x)| = \sqrt{(f(x))^2}$$ Тогда $$(|f(x)|)' = \frac{1}{2\sqrt{(f(x))^2}}*2f(x)*(f(x))'=\frac{1}{|f(x)|}*f(x)*(f(x))' = sign (f(x))*f'(x)$$

Получаем

$$sign(y_i - (w_0 + w_1 * x_i)) * (y_i - (w_0 + w_1 * x_i))'$$

Производная 

$$\frac {dQ}{dw_1} = -x_i$$

$$\frac {dQ}{dw_0} = -1$$

т.к. это производная сложной функции функционал ошибки которому на вход подаеться сама функция w_0 + w_1 * x_1 нужно так же домножить на производную этой функции

Итого получаем в случае с w_1

$$sign(y_i - (w_0 + w_1 * x_i)) * -x_i^2$$

В случае с w_0

$$sign(y_i - (w_0 + w_1 * x_i)) * -x_i$$
